## Here we do some experiments regarding the AKS primality test

```python
def check_identity(m,a):
    ans=[math.comb(m,k)*a**k for k in range(m+1)]
    ans_reduced=[number%m for number in ans]
    return ans, ans_reduced
```

```python
print(check_identity(97,6)[1])
```

## Check primality

```python
def isprime(n):
    if n == 2:
        return True
    for i in range(2, math.floor(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

```python
def count_primes(n):
    ans=0
    for i in range(2,n+1):
        if isprime(i)==True:
            ans=ans+1
    return ans
```

```python
for i in range(5,100):
    print(count_primes(i),i/math.log(i))
```

### plot primes counts and approximations

```python
def plot_prime_counts(N):
    numeros=list(range(2,N))
    y = [count_primes(i) for i in numeros]
    x = numeros

    # Plot the curve
    plt.plot(x, y,color='red', label='Prime count')
    
    Y = [m/math.log(m,2) for m in numeros]
    X = numeros

    # Plot the curve
    plt.plot(X, Y,color='blue', label='Dirichlet bound')
    # Add title and labels
    Z = [m/math.log(m) for m in numeros]
    W = numeros

    # Plot the curve
    plt.plot(W, Z,color='green', label='Prime number bound')
    plt.title("Prime count approximations")
    plt.xlabel("N")
    plt.ylabel("Time in seconds")
    plt.legend()
    #plt.xlim(a,b)

    # Display the plot
    plt.show()
```

```python
plot_prime_counts(10000)
```

### Fast exponentiation

```python
def fast_exp(a, b):
    result = 1
    exp=b
    base = a
    
    while exp > 0:
        if exp % 2 == 1:
            result = result*base
        base = base**2
        exp //= 2
        
    return result
```

```python
A=15
B=24
fast_exp(A,B)-A**B
```
