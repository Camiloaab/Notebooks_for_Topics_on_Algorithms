## Experiment to verify that primes equidistribute on congruence classes

```python
import math
```

```python
def trivial(n):
    if n == 2:
        return True
    for i in range(2, math.floor(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

```python
def findprimes(N):
    ans=[]
    for i in range(2,N+1):
        if trivial(i):
            ans.append(i)
    return ans
```

```python
def countbycongruence(N,p):
    primos=findprimes(N)
    C=dict()
    for i in range(0,p):
        C[i]=len([x for x in primos if x%p==i])
    return C
```

```python
C=countbycongruence(100000,11)
```
